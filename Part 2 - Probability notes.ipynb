## Probablity

* Probability is likelihood of an event occuring
* Probability can be measured with numeric values from 0 till 1
* Formula for probability:
<br>
$$ P(X) = \frac{Preferred \mspace{9mu} outcomes}{Sample \mspace{9mu} space} $$

X - event
<br>
P(X) - probability

Examples
* Fliping coin (1 prefered outcome and 2 possible outcomes)

$$ P(X) = \frac{1}{2} = 0.5 $$

* Roling dice (1 prefered outcome and 6 possible outcomes)

$$ P(X) = \frac{1}{6} = 0.167 $$

* Roling dice anf getting number divisible by 3 (get numbers 3 or 6) (2 prefered outcome and 6 possible outcomes)

$$ P(X) = \frac{2}{6} = 0.33 $$

* Getting two probabilities at the same time is product of each event happening independent (probablity of getting ace of dimond is equal to probabilities of getting ace times probability getting dimond)

$$ P(Ace of dimond) = P(Ace) . P(dimond) $$

**Expected values **(example with coin flip)
* Flip coin once - trial
* Flip coin 20 times - multiple trails = experiment
* Coin toss 20 times - record 20 outcomes = single experiment with 20 trails
* When we dont know theoretical (true) probabilities we need to do experimental probabilities

* Experimental prbabilities:
<br>
$$ P(X) = \frac{successful \mspace{9mu} trials}{all \mspace{9mu} trials} $$

* Expected value E(X) - outcome we expect to occur when we run an experiment
* Expected value for categorical value:
<br>
$$ P(X) = n \times p $$ 

* Expected value for numeric variables:
<br>
$$ P(X) = \sum_{i=1}^{n} x_{i} \times p_{i} $$

* Example - you have four aces, to calculate expected value of getting ace of dimond if we pick card from 4 cards 20 times we need to multiply probabilty of picking dimond to 20
<br>
$$ E(X) = P(ace \mspace{9mu} of \mspace{9mu} dimond) * n =  0.25 * 20 = 5 $$
<br>
we can read result as - if we pick one of four cards, look at it, put back and shuffle we can expect 5 times to see ace of dimonds

* Frequency - how often result can occur

* Compliment of event is everything event is not (marked with apostrophe) $$ A^{'}$$ 
* Probability of complimnet:
$$ P(X^{'}) = 1 - P(X) $$

* Sum of all prbabilites should be 1 (100% certain:
$$ P(A) + P(B) + P(C) = 1 $$




**Permutations**

* Permutations represent number of possible ways we can arrange a number of elements
* Permutations formula:
$$ P(n) = n \times (n-1) \times (n-2) \mspace{9mu} \times ... \times \mspace{9mu} 1 = n! $$
* n! - n factorial

<br>
Example - if we need arrange five people, we would have P(5) = 120

* Negative numbers don't have a factorial
* 0! = 1
* Formula:
$$ n! = (n-1) \times n $$

**Variations**

* Formula:
$$ \bar{V}_{n}^{p} = n^{p} $$
<br>
n - total number of elements available
<br>
p - number of positions which needs to be filled
<br>

**Variation without repetitions**

* Formula:
$$ V_{n}^{p} = \frac{n!}{(n-p)!} $$

**Combinations**

* Formula for combinations without repetitions:
$$ C(n, p) = \frac{n!}{(n-p)! \times p!} $$

* Formula for combinations with repetitions:
$$ \bar{C}(n, p) = \frac{(n+p-1)!}{(n-1)! \times p!} $$

**Sets and events**

* Set is a collection of elements, which hold certain values
<br>
<br>
Notations:
<br>
* Null set or empty set is denoted $$ \, \varnothing $$
* Element x is part of set A
$$ x \in A $$

* Set A contains element x
$$ A \ni  x $$

* Element x is not part of set A
$$ x \notin A $$

* For all/ any x such that...
$$ \forall x: $$

* For all elements of x in set A, such that, x is even
$$ \forall x \in A: x \mspace{9mu}is\mspace{9mu} even $$

* A is subset of B
$$ A \subseteq B $$

* A intersect B
$$ A \cap B $$

* A union B
$$ A \cup  B $$

* Some examples
 * If A and B don't intersect
 $$ A \cup  B = A + B $$
 $$ A \cap  B = \, \varnothing $$
 * If A and B have intersect
 $$ A \cup  B = A + B - A \cap B $$
 * If B is subset of A
 $$ A \cap  B = B $$
 $$ A \cup  B = A $$

* If likelihood of event A occuring (P(A)) is affected event B occuring, then A and B are dependent events (if not than they are independent)
$$ P(A \mid B) $$

* Additive law
 $$ P(A \cup  B) = P(A) + P(B) - P(A \cap B) $$

 * Multiplication rule
 $$ P(A \cap  B) = P(A \mid B) \times P(B) $$

 * Bayes law
 $$ P(A \mid B) = \frac{P(B \mid A) \times P(A)}{P(B)} $$

**Probability distributions**

* Distribution shows possible values a random variable can take and how frequently they occur

* Notation
  * Y - actual outcome
  * y - one of possible outcomes
  * $$ P(Y=y) = p(y) $$
  * Mean population - $$ \mu $$
  * Mean sample - $$ \bar{x} $$
  * Variance population $$ \sigma^{2} $$
  * Variance sample $$ s^{2} $$
  * Standard deviation for population $$ \sigma $$
  * Standard deviation sample $$ s $$
  <br>
* Discrete distributions have finitely many different possible outcomes

* Uniform distribution - U(a,b) => all outcomed are equally likely (expected value and variance have no predictive power

* Bernoulli distribution - Bern(p) => consists of single trial with two possible outcomes (example coin toss)

* Binomial distributions - B(n,p) => sequence of identical Bernouli events

* Poisson distributions - when we want to know likelihood of certain event occuring over a given interval of time or distance. Notation:
$$ Po(\lambda ) $$

* Continuous distributions - if the possible values a random variable can take are a sequence of infinitely many consecutive values.
 * PDF - probability density function
 * PDC - probability distribution curve
 * CDF - cumulative distribution function

* Normal distribution - represents a distribution that most natural events fololow (bell-shaped graph, which is symmetric and has thin tails). Notation:
$$ N(\mu, \sigma^{2}) $$

* Standardizing a normal distribution - to standardize any normal distribution we need to transform it so that the mean is 0 and the variance and standard deviation are 1. After normalizing we can use Z-score table or simply Z-table
$$ z = \frac{y - \mu}{\sigma} $$

* Students T distribution - same as normal distribution but with small saple size (ussualy have fatter tails than normal distribution). Notation - t(k) where k is degree of freedom

* Chi-squared distribution - asymetric, starts with 0 and graph is skewd to right side
$$ \chi^{2} (k) $$

* Exponential distribution - usually observed in events which significaly change early on. Notation
$$ Exp(λ) $$

* Continuos logistic distribution is observed when trying to determine how continuous variable inputs can affect the probability of binary outcome. Notation
$$ Logistic(μ, s) $$
